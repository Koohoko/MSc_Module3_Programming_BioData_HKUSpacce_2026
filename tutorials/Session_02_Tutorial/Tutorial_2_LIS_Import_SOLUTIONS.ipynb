```python
# 1. Define the URL
url <- "https://raw.githubusercontent.com/Koohoko/MSc_Module3_Programming_BioData_HKUSpacce_2026/main/data/mock_patient_data.csv"

# 2. Read the data
lis_data <- read.csv(url)

# 3. Check sample receipt
print("Data Dimensions:")
print(dim(lis_data))
head(lis_data)
```

## Step 2: Data Inspection & "Sanity Checks"

**Task:** Verify the data types. Are Ages numeric? Is Status text?
Look at the summary for Age. Is there a Min or Max that looks impossible?

```python
# Check structure
str(lis_data)

# Summary statistics (Quick QC)
summary(lis_data)
```

## Step 3: Extracting Vectors for Analysis

**Task:** Calculate the median Viral Load of the batch.

```python
# Pull out the Viral_Load column as a vector
vl_vector <- lis_data$Viral_Load

# Calculate statistics
median_vl <- median(vl_vector)
max_vl <- max(vl_vector)

print(paste("Median Viral Load:", median_vl))
print(paste("Highest Viral Load:", max_vl))
```

## Step 4: Logical Filtering (The "Critical List")

**Scenario:** The doctor wants a list of **Male** patients who are **Positive**.
**Task:** Create a complex subset.

```python
# 1. Define the criteria
# Gender is Male AND Status is Positive
criteria <- (lis_data$Gender == "M") & (lis_data$Status == "Positive")

# 2. Apply the filter (Row selection, All columns)
critical_cases <- lis_data[criteria, ]

# 3. View the result
print("Critical Cases List:")
print(critical_cases)
```

## Step 5: Exporting the Report

**Task:** Save the `critical_cases` to a CSV file named `critical_cases_report.csv`.

```python
write.csv(critical_cases, "critical_cases_report.csv", row.names = FALSE)

# Verify it exists
list.files()
```
