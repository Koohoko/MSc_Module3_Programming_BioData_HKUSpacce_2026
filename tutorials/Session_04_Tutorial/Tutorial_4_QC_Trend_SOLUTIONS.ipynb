## Step 1: Load Data

```python
url <- "https://raw.githubusercontent.com/Koohoko/MSc_Module3_Programming_BioData_HKUSpacce_2026/main/lectures/Session_04_Visualization/data/daily_qc_trend.csv"
qc_data <- read.csv(url)
qc_data$Date <- as.Date(qc_data$Date)
head(qc_data)
```

## Step 2: Basic Time Series (Color by Instrument)

```python
p <- ggplot(qc_data, aes(x = Date, y = Result, color = Instrument)) +
  geom_line() +
  geom_point()
p
```

## Step 3: Facet by Instrument

```python
p <- p + facet_wrap(~ Instrument)
p
```

## Step 4: Add QC Limits

```python
p <- p +
  geom_hline(yintercept = 5.0, color = "black") +
  geom_hline(yintercept = 5.4, color = "red", linetype = "dashed") +
  geom_hline(yintercept = 4.6, color = "red", linetype = "dashed")
p
```

## Step 5: Final Polish

```python
p <- p +
  theme_bw() +
  labs(title = "Monthly Levey-Jennings Chart",
       y = "Glucose Level (mmol/L)",
       x = "Date")
p
```
