```python
# 1. Define the URL
url <- "https://raw.githubusercontent.com/Koohoko/MSc_Module3_Programming_BioData_HKUSpacce_2026/main/lectures/Session_03_Logic_and_Control/data/hba1c_results.csv"

# 2. Read the data
df <- read.csv(url)

# 3. Inspect the first few rows
head(df)
```

## Step 2: Initialize Storage

**Task:** We need a place to put our diagnoses. Create a new column named `Diagnosis` in the dataframe and fill it with `NA` (Not Available) as a placeholder.

```python
# Create a new column filled with NA
df$Diagnosis <- NA

# Check structure to confirm the column exists
head(df)
```

## Step 3: The Logic (Mental Draft)

Before coding, verify the logic:
- IF val >= 6.5 -> "Diabetes"
- ELSE IF val >= 5.7 -> "Prediabetes"
- ELSE -> "Normal"

## Step 4 & 5: The Loop & Logic

**Task:** Write a `for` loop that iterates through every row of the dataframe. Inside the loop, check the `HbA1c_Level` and assign the correct string to the `Diagnosis` column.

```python
# Loop from 1 to the number of rows
for (i in 1:nrow(df)) {

   # 1. Get the value for this specific row
   val <- df$HbA1c_Level[i]

   # 2. Apply Logic (if / else if / else)
   if (val >= 6.5) {
      df$Diagnosis[i] <- "Diabetes"
   } else if (val >= 5.7) {
      df$Diagnosis[i] <- "Prediabetes"
   } else {
      df$Diagnosis[i] <- "Normal"
   }
}
```
